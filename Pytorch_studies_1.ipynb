```python
## Creating Tensors 

# Scalar
print("Scalar")
scalar = torch.tensor(7) # singlue number is printed
print(scalar)
print(scalar.ndim)
print(scalar.item()) # retrieve the number from tensor : torch.tensor -> python int
print()

# Vector 
print("Vector")
vector = torch.tensor([7,7])
print(vector)
print(vector.shape)
print()

# Matrix 
print("Matirx")
matrix = torch.tensor([[7,8],
                       [9,10]])

print(matrix.ndim)
print(matrix.shape)
```

```python
# Create a random tensor of size (3,4)

random_tensor = torch.rand(size = (3,4))
print(random_tensor)
print()
print(random_tensor.dtype)
print()
print(random_tensor.shape)
```

```python
## Creating Tensors with Zeros and Ones 
```

```python
print("Zeros")
zeros = torch.zeros(size = (3,4))
print(zeros)
print()
print(zeros.dtype)
print()
print(zeros.shape)
print()

print("Ones")
ones = torch.ones(size = (3,4))
print(ones)
print()
print(ones.dtype)
print()
print(ones.shape)
```

```python
## Tensor Data types 
```

```python
float_32_tensor = torch.tensor([3.0,6.0,9.0],
                               dtype = None,
                               device = None,
                               requires_grad = False)

float_16_tensor = torch.tensor([3.0,6.0,9.0],
                               dtype = torch.float16)

print("32")
print(float_32_tensor)
print(float_32_tensor.dtype)
print(float_32_tensor.device)
print()
print("16")
print(float_16_tensor)
print(float_16_tensor.dtype)
print(float_16_tensor.device)
```

```python
## Manipulating Tensor 
tensor = torch.tensor([1,2,3])
tensor + 10
print(tensor+10)
print(torch.add(tensor,10))
```

```python
# Addition 
tensor * 10
print(tensor * 10)
print(torch.multiply(tensor,10))
```

```python
# Subtraction 
tensor-10
```

```python
# Multiplication(Elment-wise)
print(tensor * tensor)
```

```python
# Division
print(tensor/2)
```

```python
# Matrix Multiplication 
# torch.matmul()
# @ 
# torch.mm() -> short cut for matmul 

print(torch.matmul(tensor,tensor)) 
print(tensor @ tensor)
```

```python
## Transpose 
```

```python
tensor_A = torch.tensor([[1, 2],
                         [3, 4],
                         [5, 6]], dtype=torch.float32)

tensor_B = torch.tensor([[7, 10],
                         [8, 11], 
                         [9, 12]], dtype=torch.float32)

torch.matmul(tensor_A, tensor_B) # (this will error)
```

```python
print(tensor_A)
print(tensor_B)
```

```python
print(tensor_A)
print(tensor_B.T)
```

```python
print(torch.mm(tensor_A,tensor_B.T))
```

```python
## Finding the min,max,mean,sum,etc(aggregation)
```

```python
x = torch.arange(0,100,10)
print(x)
print(x.max())
print(x.min())
print(x.type(torch.float32).mean()) # You may find some methods such as torch.mean() require tensors to be in torch.float32 (the most common) or another specific datatype, otherwise the operation will fail.
print(x.sum())
```

```python
# positional min,max(argmin, argmax )
```

```python
print(x.argmax())
print(x.argmin())
# Returns index of max and min values 
```

```python
## Reshaping, stacking, squeezing and unsqueezing 
```

```python
# torch.reshape()
x = torch.arange(7)
print(x)
x.reshape(1,7)
print(x)
```

```python
# torch.view()
x = torch.arange(7)
print(x)
x.view(1,7)
print(x)
```

```python
# stack
x_stacked_dim_0 = torch.stack([x,x,x,x],dim = 0)
print(x_stacked_dim_0)
print()
x_stacked_dim_1 = torch.stack([x,x,x,x],dim = 1)
print(x_stacked_dim_1)
```

```python
# squeeze 

x = torch.ones(2,2,1,3)
print(x)
print()
print(x.squeeze())
print()
print(x.squeeze().shape)
print()
print(x.squeeze(dim= 2))
print()
print(x.squeeze(dim= 2).shape)
```

```python
x = torch.ones(2,2)
print(x)
print()
print(x.unsqueeze(dim=0))
print()
print(x.unsqueeze(dim=0).shape)
print()
print(x.unsqueeze(dim=1))
print()
print(x.unsqueeze(dim=1).shape)
print()
print(x.unsqueeze(dim=2))
print()
print(x.unsqueeze(dim=2).shape)
```
